```python
import numpy as np

from src.eval.bootstrap import bootstrap_metric
from src.eval.delong import delong_test

rng = np.random.default_rng(3)
labels = rng.integers(0, 2, size=500)
scores_a = rng.random(500)
scores_b = rng.random(500)

bootstrap_auc = bootstrap_metric(labels, scores_a, lambda y, s: float((y * s).mean()), n_resamples=100)
print('Bootstrap estimate:', bootstrap_auc)

z, p = delong_test(labels, scores_a, scores_b)
print(f'DeLong z={z:.3f}, p={p:.3f}')
```
